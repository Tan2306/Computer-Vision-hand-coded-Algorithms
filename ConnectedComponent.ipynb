```python
def gray2binary(I,th):
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            if I[i,j]>=th:
                I[i,j]=1
            else:
                I[i,j]=0
    return I           
                
def bwconnected(BW,conn):
    
    componentno=0
    componentmatrix=np.zeros([int(BW.shape[0]),int(BW.shape[1])])
    
    for i in range(0,BW.shape[0],1):
        for j in range(0,BW.shape[1],1):
            # for each pixel in BW
            
            print()
            print("new pixel",i,j)
            print("pixel value=",image[i,j])
            
            if conn==4:
                neighbours=[(i-1,j),(i+1,j),(i,j-1),(i,j+1)]
            if conn==8:
                neighbours=[(i-1,j),(i+1,j),(i,j-1),(i,j+1),(i-1,j-1),(i+1,j+1),(i-1,j+1),(i+1,j-1)]
            
            for k in neighbours: # for boundary pixels
                if k[0]<0 or k[0]>BW.shape[0]-1 or k[1]<0 or k[1]>BW.shape[1]-1:
                    neighbours.remove(k)
                    
            print("neighbours=",neighbours)
            
            
            if BW[i,j]==0:# -1 signifies (in the component matrix) that pixel has been traversed
                componentmatrix[i,j] = -1
                print("for this pixel 0, component matrix value is = ",componentmatrix[i,j])

            if BW[i,j]==1:#Either a new component or a part of the old one
                neighbours_component=[] #list containing the component matrix values of the neighbours
                for k in neighbours:
                    neighbours_component.append(componentmatrix[k[0],k[1]])
                print("neighbours_component=",neighbours_component)              
                
                for x in neighbours_component:#if one in neighbour has a component, it takes from there
                    if x!=0 and x!=-1:
                        print("x=",x)
                        componentmatrix[i,j]=x
                        
                if componentmatrix[i,j]==0:#else, new component is made
                    print("compnent no. now=",componentno)
                    componentno=componentno+1
                    componentmatrix[i,j]=componentno
                    
                print("for this pixel 1 , component matrix value is = ",componentmatrix[i,j])
                
    print("value of i and j is ",i,j)
    return componentno, componentmatrix
                    
            
    
image=cv2.imread("birds_gray.tif",-1)
image=image[0:30,0:30]

BW=gray2binary(image,90)

cv2.imshow("BW",BW)
cv2.waitKey(0)
cv2.destroyAllWindows()


CC,componentmatrix=bwconnected(BW,4)
print()
print()
print("Number of components=",CC)
unique, counts = np.unique(componentmatrix, return_counts=True)

for i in range(len(counts)):
    if counts[i]==max(counts):
        print("The largest component is component number : ",unique[i])
```
